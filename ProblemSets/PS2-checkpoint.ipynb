```python
from sympy import *
import sympy as sy
import numpy as np
import matplotlib.pyplot as plt
import math

# define the function f1 as written in the ACME text
def f1(x):
    return (sy.sin(x)+1)**sy.sin(sy.cos(x))
# takes the function and returns the symbolic derivative
def f1_prime(x):
    return sy.diff(f1(x), x)
# Lambdify the function to accept NumPy arrays
f_lam = sy.lambdify((x), f1(x), "numpy")
fprime_lam = sy.lambdify((x), f1_prime(x), "numpy")

# Check the functions by plotting them as described in the text
x_values = np.linspace(-math.pi, math.pi, 5000)
plt.plot(x_values,f_lam(x_values), label="f(x)")
plt.plot(x_values,fprime_lam(x_values), label="f'(x)")
plt.legend()
plt.title("Problem 1")
plt.show()
```

```python
# Six functions for the 6 finite difference quoitents from the text
# each takes function handle, f, an array of points x, and a float h
# each returns an array of difference quotients evaluated at each point

def forward_1(f, x, h):
    return (f(x+h) - f(x)) / h

def forward_2(f, x, h):
    return (-3*f(x) + 4*f(x+h) - f(x+2*h) ) / (2*h)

def backward_1(f, x, h):
    return (f(x) - f(x-h)) / h

def backward_2(f, x, h):
    return (3*f(x) - 4*f(x-h) + f(x-2*h)) / (2*h)

def centered_2(f, x, h):
    return (f(x+h) - f(x-h)) / (2*h)

def centered_4(f, x, h):
    return (f(x-2*h) - 8*f(x-h) + 8*f(x+h) - f(x+2*h)) / (12*h)

# Testing the functions by approximating the derivative

h=0.01

plt.plot(x_values, fprime_lam(x_values), label="f’(x)")
plt.plot(x_values, forward_1(f,x_values,h), label="forward-1")
plt.plot(x_values, forward_2(f,x_values,h), label="forward-2")
plt.plot(x_values, backward_1(f,x_values,h), label="backward-1")
plt.plot(x_values, backward_2(f,x_values,h), label="backward-2")
plt.plot(x_values, centered_2(f,x_values,h), label="centered-2")
plt.plot(x_values, centered_4(f,x_values,h), label="centered-4")
plt.legend()
plt.title("Problem 2")
plt.show()
```
